```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# 1) Generate a 100x100 matrix M of random numbers according to the uniform distribution U[0, 1]
M = np.random.uniform(0, 1, (100, 100))

# 2) Choose one single column xi of the matrix M and consider its normalization xei
column_index = np.random.randint(0, 100)  # Choose a random column index
xi = M[:, column_index]
mean_xi = np.mean(xi)
std_dev_xi = np.std(xi)
xei = (xi - mean_xi) / std_dev_xi

# 3) Build the series of averages of all columns xa
xa = np.mean(M, axis=0)

# Plot histograms for both the normalized column xei and the series of averages xa
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(xei, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram of Normalized Column')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(xa, bins=20, alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram of Series of Averages')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot normal distribution curve over the histogram of xa
mean_xa = np.mean(xa)
std_dev_xa = np.std(xa)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_xa, std_dev_xa)
plt.plot(x, p, 'r', linewidth=2)


plt.tight_layout()
plt.show()


```
