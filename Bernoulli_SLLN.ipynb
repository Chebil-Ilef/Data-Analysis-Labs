```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers according to Bernoulli distribution with p = 1/2
p1 = 1/2
X1 = np.random.binomial(1, p1, size=100) #n=1 from a binomial= Bernoulli.
series_half = np.cumsum(X1) / np.arange(1, 101)

# Generate random numbers according to Bernoulli distribution with p = 1/3
p2= 1/3
X2 = np.random.binomial(1, p2, size=100)
series_third = np.cumsum(X2) / np.arange(1, 101)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, 101), series_half, label='p = 1/2')
plt.plot(np.arange(1, 101), series_third, label='p = 1/3')
plt.ylabel('Series Value')
plt.title('Bernoulli Distribution with Different Parameter P')
plt.legend()
plt.grid(True)
plt.show()
```
